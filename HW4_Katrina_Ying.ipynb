```python
#    1c. Most expensive trip (total amount)
# if we want to show the row not just the amount
# A different way of accomplishing the same task with a subquery.

q = """
  SELECT *
  FROM nyc
  WHERE Total_amount = (
      SELECT MAX(Total_amount)
      FROM nyc
  );
    """
db.sql(q)
```

```python
## 2a. Most expensive trip per mile (total amount/mile)

q = """
  SELECT MAX(Total_amount/Trip_distance) AS 'most expensive trip per mile'
        FROM nyc;

"""

db.sql(q)
```

```python
##    2b. Most expensive trip per mile (total amout/mile)
# if we want to show the row not just the amount


q = """
  SELECT *
  FROM nyc
  WHERE Total_amount / Trip_distance = (
    SELECT MAX(Total_amount/Trip_distance) AS 'most expensive trip per mile'
    FROM nyc
  );

"""
db.sql(q)


## Notice the quote of Max Dollar/Mile in WHERE and MAX
```

```python
##   3a. Most generous trip (highest tip)

q = """
  SELECT MAX(Tip_amount) AS Max_tip_amount
  From nyc
    """
db.sql(q)
```

```python
##   3b. Most generous trip (highest tip)
# if we want to show the row not just the amount
# Let's rename the tip amount as 'Most Generous Tip'

q = """
SELECT *, Tip_amount AS 'Most Generous Tip'
FROM nyc
ORDER BY Tip_amount DESC
LIMIT 1
    """
db.sql(q)

## Notice the quote of Most Generous Tip in ORDER BY
```

```python
##    4. Longest trip duration
## use strftime(): This function is used to format a datetime value based on a specified
## format. Calculate the minutes-equivalent of each datetime (dropoff and pickup),
## Then cast these minutes values to integers
## Subtract to find the duration
## Get the max

q = """ SELECT MAX(strptime(Lpep_dropoff_datetime, '%Y-%m-%d %H:%M:%S') - strptime(lpep_pickup_datetime, '%Y-%m-%d %H:%M:%S' )) from nyc"""

db.sql(q)
```

```python
##    5. Mean tip by hour


q = """
  SELECT
    CAST(SUBSTR(lpep_pickup_datetime, 12, 2) AS INTEGER) AS pickup_hour,
    AVG(Tip_amount) AS mean_tip
FROM
    nyc
GROUP BY
    SUBSTR(lpep_pickup_datetime, 12, 2)
ORDER BY
    pickup_hour;


    """
db.sql(q)
```

```python
###    7. Average passenger count per trip. Use ROUND() function if you need to.

q = """
  SELECT
  ROUND(AVG(passenger_count), 2) AS avg_pc
  FROM nyc
    """
db.sql(q)
```

```python
#   8. Average passenger count per trip by hour

q = """
  SELECT
    CAST(SUBSTR(Lpep_pickup_datetime, 12, 2) AS INTEGER) AS pickup_hour,
    AVG(passenger_count) AS avg_pc
  FROM
    nyc
  GROUP BY
    SUBSTR(Lpep_pickup_datetime, 12, 2)
  ORDER BY
   pickup_hour
    """
db.sql(q)
```

```python
##  9a. Which airport welcomes more passenger: JFK or EWR?
## Tip: check RateCodeID from data dictionary for the definition (2: JFK, 3: Newark)

q = """
  SELECT
    SUM(IF(RateCodeID = 2, passenger_count, 0)) AS JFK_p,
    SUM(IF(RateCodeID = 3, passenger_count, 0)) AS EWR_p,
    FROM
      nyc
    WHERE
      RateCodeID in (2, 3)
    """
db.sql(q)

```

```python
## 9b. Use a CASE statement to re-code RateCodeID=2 as JFK, and 3 as EWR

q = """
  SELECT
    CASE
      WHEN RateCodeID = 2 THEN 'JFK'
      WHEN RateCodeID = 3 THEN 'EWR'
    END AS airport,
    SUM(passenger_count) AS total_p
  FROM
    nyc
  WHERE
    RateCodeID IN (2, 3)
  GROUP BY
    RateCodeID

    """
db.sql(q)

## We recodded 2 and 3 and JFK and EWR
```
